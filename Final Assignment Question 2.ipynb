```python
## 2. Keep Approximating
## Part (b,c,d)

f = tan(x)
f_1 = diff(f)
f_2 = diff(f_1)
f_3 = diff(f_2)

L(x) = f(0) + x * f_1(0) 
Q(x) = f(0) + x * f_1(0) + (x^2/2) * f_2(0)
C(x) = f(0) + x * f_1(0) + (x^2/2) * f_2(0) + (x^3/6) * f_3(0)
```
