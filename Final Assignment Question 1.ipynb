```python
sawtooth(x) = (x - floor(x))
plot(sawtooth, (-5,5)).show(ymin = -1, ymax = 2)
```

### 1.c. Limit Calculation

```python
plot1 = plot(sgn(x+2), -7,3, axes_labels=['$x$','$f(x)=sgn(x+2)$'])
p = point((-1.95,1),color='red',size=20)
plot1+p
```

```python
lim(sgn(x+2), x=-2, dir='+')
```

```python
plot1=plot(sgn(x+2)**2, -7,3, axes_labels=['$x$','$f(x)=sgn^2(x+2)$'])
p =point((-2,1),color='red',size=20)
plot1+p
```

### 1.d. Sketching the function and limit at even integer

```python
f(x) = (2*sawtooth + (-1)**floor(x))/2
plot1 = plot(f,-5,5, axes_labels=['$x$','$f(x)$'])

a = [-4,-2,0,2,4]
p = [0,0,0,0,0,0]

for i in range(5):
    p[i] = point((a[i],0.5),color='red',size=20)
plot1+p[1]+p[2]+p[3]+p[0]+p[4]
```

```python
limit(f,x=4)
```

### 1.e. Continuous Function

```python
p(x) = piecewise((((-oo, -1), x/abs(x)),([-1,1], (2*(x - floor(x)) + (-1)^(floor(x)))/2 - .5),((1,oo), x/abs(x))))
plot(p, (-7,7)).show(ymin = -1, ymax = 1)
```

### 1.f. Discontinuous at multiples of 3 and 7

```python
p(x) = ((x/7 - floor(x/7)) + (-1)^(floor(x/3)))
plot(p, (-20,20)).show(ymin = -5, ymax = 5)
```
